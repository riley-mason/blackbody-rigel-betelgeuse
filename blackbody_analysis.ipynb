# Blackbody Radiation: Rigel vs. Betelgeuse

This notebook computes and visualizes the blackbody energy distribution for two stars in Orion:
- **Rigel** (blue supergiant): 11,000 K
- **Betelgeuse** (red supergiant): 3,500 K

The formula used is Planck's law: B(w,T) = (2hc²/w⁵) × 1/(e^(hc/wkT) - 1)

#### XXX Programming Question  
Compute the Blackbody energy distribution for two stars and plot them.  

The energy spectrum of the Blackbody radiation is given by: 

B(w, T) = (2 * h * c^2) / ((w^5) * (exp(h * c / w * k * T) - 1))  

w = wavelength in meters  

T = temperature in kelvins. 

h = Planck's constant  

c = speed of light  

k = Boltzmann's constant  

The two stars are Rigel and Betelguese in the constellation Orion.

Rigel's Temperature = 11,000 K  

Betelguese's Temperature = 3,500 K

```python
# Import libraries
import numpy as np          # for array operations and numerical methods  
import matplotlib.pyplot as plt  # for plotting
from scipy.constants import h, c, k  # access physical constants in SI units (Planck's const, speed of light, Boltzmann const)
```

```python
# Complete the function that takes as input wavelength in nano-meters and temperature in kelvins
# Converts the wavelength to meters and returns the energy radiated in watts per square meter
def blackbody (w, T):
    # Convert wavelength from nanometers to meters
    w = w * 10**-9
    
    # Planck's law: B(w,T) = (2hc²/w⁵) × 1/(e^(hc/wkT) - 1)  
    B = (2 * h * c**2) / ((w**5) * (np.exp((h * c) / (w * k * T)) - 1))
    return B
```

```python
# Generate a numpy array of 1000 points that spans from 100 nm to 3000 nm 
# Range covers UV (100 nm) through visible to near-IR (3000 nm)
w = np.linspace(100,3000,1000)
```

```python
# Compute the blackbody energy distribution for Rigel
# Rigel is a blue supergiant with surface temperature ~11,000 K
Rigel = blackbody(w, 11000)
```

```python
# Compute the blackbody energy distribution for Betelguese
# Betelguese is a red supergiant with surface temperature ~3,500 K
Betelguese = blackbody(w, 3500)
```

```python
# Plot Energy Distribution
# Use log scale on y-axis because blackbody intensity spans many orders of magnitude


plt.semilogy(w, Rigel, color="blue", label="Rigel")
plt.semilogy(w, Betelguese, color="red", label="Betelguese")
plt.axvspan(380, 750, alpha=0.1, color='yellow', label='Visible spectrum')
plt.xlabel("Wavelength (nanometers)")
plt.ylabel("Energy in watts per square meter")
plt.title("Blackbody Distribution for Rigel and Betelguese")
plt.legend()
```
